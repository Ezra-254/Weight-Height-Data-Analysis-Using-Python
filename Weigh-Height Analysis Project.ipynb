# BMI categories:
## Underweight: BMI less than 18.5 
## Normal weight: BMI between 18.5 and 24.9 
## Overweight: BMI between 25 and 29.9 
## Obese: BMI of 30 or more

```python
def bmi_cat(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'
```

```python
df['BMI Category'] = df['BMI'].apply(bmi_cat)
```
